# ordering the Frame

```python
import cv2
import dlib
import numpy as np
import os
from imutils import paths
import pickle
import random
from scipy.stats import norm
```

```python
SourcePaths = list(paths.list_images('/content/data_video'))
```

```python
SourcePaths
```

```python
imagePaths = list(map(lambda x: x.split('/')[-1], SourcePaths))
```

```python
user_id  = list(map(lambda x: x.split('/')[4], SourcePaths))
```

```python
image_id = list(map(lambda x: x.split('_')[-1].replace('.jpg',''), imagePaths))
```

```python
for (_, id) in enumerate(image_id):
  if 'test' in id:
    image_id[_] = id.replace('test','')
```

```python
image_id = list(map(lambda x: int(x), image_id))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list(zip(SourcePaths, imagePaths, image_id, user_id)), columns=['src', 'path', 'id', 'user_id'])
```

```python
df
```

# Client

```python
client_mask = list(map(lambda x: True if '/client/' in x else False, df['src'].values))
```

```python
client = df[client_mask]
```

```python
client
```

```python
client_dict = {}
for uId in np.unique(client['user_id']):
  client_dict[uId] = client[client['user_id'] == uId].sort_values(by=['id']).drop(['path', 'id', 'user_id'], axis=1)
```

```python
client_all = None
for (key, dataf) in client_dict.items():
  if client_all is None:
    client_all = dataf.values
  else:
    client_all = np.vstack((client_all, dataf.values))
```

```python
client_all = pd.DataFrame(client_all)
```

```python
client_all.to_csv('client.txt', index=False,header=False)
```
