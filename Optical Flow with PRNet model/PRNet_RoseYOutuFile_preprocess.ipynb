```python
import sys
import dlib
import cv2
import numpy as np
import imutils
import numpy as np
import os
import tqdm
```

```python
import matplotlib.pyplot as plt
```

```python
for video in tqdm.tqdm(videoPaths):
  folder = video.split('/')[-1].replace('.mp4','')
  if not os.path.exists(f'/content/data/imposter/{folder}'):
    os.makedirs(f'/content/data/imposter/{folder}')
  # /content/data/imposter
  cam = cv2.VideoCapture(video)
  detector = dlib.get_frontal_face_detector()
  predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
  i = 0
  count = 0
  while cam.isOpened():
      ret, frame = cam.read()
      
      if ret:
        dets = detector(frame)
        if dets:
            max_a = (dets[0].bottom() - dets[0].top()) * (dets[0].right() - dets[0].left())
            d = dets[0]
            if len(dets) > 1:
                for dz in dets[1:]:
                    h = (dz.bottom() - dz.top()) * (dz.right() - dz.left())
                    if h > max_a:
                        d = dz
            shape = predictor(frame, d)
        # for d in dets:
        #     shape = predictor(frame, d)
            # face = dlib.get_face_chip(
            # frame, shape, size=INPUT_SIZE)
            face = frame[d.top():d.bottom(), d.left():d.right()]
            try:
              face = cv2.resize(face, (160, 160))
              cv2.imwrite(f'/content/data/imposter/{folder}/test{i}.jpg', face)
            except Exception as e:
              print(str(e))
            i += 1
        count += 4
        cam.set(1, count)
      else:
        cam.release()
        break
```

```python
import glob
import pandas as pd
import numpy as np
```

```python
SourcePaths = glob.glob('/content/maskrose/*/*/*.jpg')
SourcePaths
```

```python
imagePaths = list(map(lambda x: x.split('/')[-1], SourcePaths))
```

```python
user_id  = list(map(lambda x: x.split('/')[4], SourcePaths))
```

```python
image_id = list(map(lambda x: x.split('_')[-1].replace('.jpg',''), imagePaths))
```

```python
for (_, id) in enumerate(image_id):
  if 'test' in id:
    image_id[_] = id.replace('test','')
```

```python
image_id = list(map(lambda x: int(x), image_id))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list(zip(SourcePaths, imagePaths, image_id, user_id)), columns=['src', 'path', 'id', 'user_id'])
```

```python
df
```

# Client

```python
client_mask = list(map(lambda x: True if '/client/' in x else False, df['src'].values))
```

```python
client = df[client_mask]
```

```python
client
```

```python
client.drop(['path', 'id', 'user_id'], axis=1).to_csv('client.txt', index=False)
```

```python
np.unique(client['user_id'])
```

```python
client[client['user_id'] == 'phuc']
```

```python
client_dict = {}
for uId in np.unique(client['user_id']):
  client_dict[uId] = client[client['user_id'] == uId].sort_values(by=['id']).drop(['path', 'id', 'user_id'], axis=1)
```

```python
client_all = None
for (key, dataf) in client_dict.items():
  if client_all is None:
    client_all = dataf.values
  else:
    client_all = np.vstack((client_all, dataf.values))
```

```python
client_all = pd.DataFrame(client_all)
```

```python
client_all.to_csv('client.txt', index=False,header=False)
```
